### 起名大师

```python
### 起名大师
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
api_key = "sk-pvdcqtmuphqjgepywixzapshexhxcugkhnrtjnuhdotlvgtx"
base_url = "https://api.siliconflow.cn/v1"
llm = OpenAI(
    model = "Qwen/Qwen2.5-7B-Instruct",
    temperature=0,
    stream=False,
    openai_api_base=base_url,
    openai_api_key=api_key,
 )
prompt = PromptTemplate.from_template("你是一个起名大师，请模仿示例起3个{county}名字,比如男孩经常被叫做{boy},女孩经常被叫做{girl}")
message = prompt.format(county="中国",boy="狗蛋",girl="翠花")
print(message)
llm.predict(message)
```

### 格式化输出

```python
from langchain.schema import BaseOutputParser
# 自定义class ，继承BaseOutputParser
class CommaSeparatedParser(BaseOutputParser):
    def parse(self, output):
        return output.split(',')
    
CommaSeparatedParser().parse("a,b,c,d,e")
```

```python
# 起名大师，输出格式为一个数组
from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

from langchain.schema import BaseOutputParser
# 自定义class ，继承BaseOutputParser
class CommaSeparatedParser(BaseOutputParser):
    def parse(self, output):
        return output.split(',')

api_key = "sk-pvdcqtmuphqjgepywixzapshexhxcugkhnrtjnuhdotlvgtx"
base_url = "https://api.siliconflow.cn/v1"
llm = OpenAI(
    model = "Qwen/Qwen2.5-7B-Instruct",
    temperature=0,
    stream=False,
    openai_api_base=base_url,
    openai_api_key=api_key,
 )
prompt = PromptTemplate.from_template("你是一个起名大师，请模仿示例起3个{county}名字,比如男孩经常被叫做{boy},女孩经常被叫做{girl}。请返回以逗号分隔的列表形式。仅返风逗号分隔的列表，不要返回其他内容。")
message = prompt.format(county="美国",boy="sam",girl="lucy")
print(message)
result = llm.predict(message)
CommaSeparatedParser().parse(result)
```
